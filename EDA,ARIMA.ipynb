# EDA

```r
df <- read.csv('/content/finalData.csv')
```

```r
df_ts <- ts(df, start = c(2012,121.25), frequency = 365.25)
```

Since, we have negative values in the dataset we are adding 50 to get positive values as adding doesnt alter any seasonality.

```r
df_ts[,'avgTemp'] <- df_ts[,'avgTemp']+50
df_ts[,'tempmax'] <- df_ts[,'tempmax']+50
df_ts[,'tempmin'] <- df_ts[,'tempmin']+50
df_ts[,'temp'] <- df_ts[,'temp']+50
```

```r
plot(df_ts[,'tempmax'], main = 'Max Temp Timeseries')
```

```r
plot(df_ts[,'temp'], main = 'Avg Temp Timeseries')
```

```r
plot(df_ts[,'tempmin'], main = 'Min Temp Timeseries')
```

Checking the data if it is stationary

```r
Acf(df_ts[,'temp'])
```

```r
Pacf(df_ts[,'temp'])
```

The ACF and PACF suggests that AR(1) model would be our best shot.

```r
kpss.test(df_ts[,'temp'])
```

```r
adf.test(df_ts[,'temp'])
```

The ADF test says that our data is not stationary.

Checking for any variance that is present in the data

```r
BoxCox.lambda(df_ts[,'temp'])
```

```r
BoxCox.lambda(df_ts[,'tempmax'])
```

```r
BoxCox.lambda(df_ts[,'tempmin'])
```

From the boxcox results we can say that there is some variance that is present and can be stabilized by uisng lambda value of 2

```r
df_ts_boxcox <- BoxCox(df_ts[,'temp'], lambda = 2)
```

```r
fit_add <- decompose(df_ts_boxcox, type = 'additive')
plot(fit_add)
```

```r
fit_mul <- decompose(df_ts_boxcox, type = 'multiplicative')
plot(fit_mul)
```

Based on the magnitude I would choose additive as the magnitude of the seasonal signal is closer to the random signal

Checking by differencing the data

```r
help('diff')
```

```r
diff_2 = diff(df_ts_boxcox, lag=365)
```

```r
plot(diff_2, main = 'After Differencing and BoxCox')
```

```r
diff_1 = diff(df_ts_boxcox, order =1)
```

```r
plot(diff_1)
```

This looks much more like a white noise

```r
kpss.test(df_ts_boxcox)
```

```r
adf.test(df_ts_boxcox)
```

Even though adf test says that data is not stationary KPPS test says that the data is stationary and the plot resembles white noise we can conclude that data is stationary hence no further differencing required for the data
